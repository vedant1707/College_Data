# Mean Method

```python
import pandas as pd
car = pd.read_csv("cars.csv", sep=";", header=[0,1])
car.fillna(car.mean())
```

# Median Method

```python
import pandas as pd
car = pd.read_csv("cars.csv", sep=";",header=[0,1])
car.fillna(car.median())
```

# Most Frequent Method

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
car = pd.read_csv("cars.csv", sep=";", header=[0,1])
imputer = SimpleImputer(missing_values=np.NaN, strategy='most_frequent')
imputer.fit(car)
mostfrequent = pd.DataFrame(imputer.transform(car))
mostfrequent
```
